```python
#"" 안에는 데이터 경로 지정, Google Colab의 경우에는 파일을 업로드해야함.
tup = uproot.open("/Users/jaewonl/lab/sitrineo/MC_root/Sr90/Sr90_0.5T_100K_Collimator_Sensor_DX10_M.cfg.root:ntp1")
tup.keys()
#['ntp1;1']

HP_df = tup.arrays(['HitsensorID', 'HitposINmmX','HitposINmmY', 'HitposINmmZ', 'HitmomMeVX','HitmomMeVY', 'HitmomMeVZ'], library='pd')

"""
treex = tup['HitposINmmX']
# Load data as a pandas DataFrame
data_dfx = treex.arrays(library='pd')

treey = tup['HitposINmmY']
# Load data as a pandas DataFrame
data_dfy = treey.arrays(library='pd')
tup.close()  # usually, it's better to open the file with a "with" statement -> needs no closing

print(data_dfx)
print(data_dfy)
"""
HP1_df = HP_df.query('HitsensorID == 0')
HP2_df = HP_df.query('HitsensorID == 1')
HP3_df = HP_df.query('HitsensorID == 2')
HP4_df = HP_df.query('HitsensorID == 3')

#Show the first 5 lines of the DataFrame
#data_df.head()

#print(HP_df)
```

```python
#Grid 정의
fig = plt.figure(figsize=(10,10))
gs = GridSpec(4, 4)

#그림 정의
ax_scatter = fig.add_subplot(gs[1:4, 0:3])
ax_hist_x = fig.add_subplot(gs[0,0:3])
ax_hist_y = fig.add_subplot(gs[1:4, 3])

#x, y, z축 변수지정
x1 = HP1_df['HitposINmmX']
y1 = HP1_df['HitposINmmY']
x2 = HP2_df['HitposINmmX']
y2 = HP2_df['HitposINmmY']
x3 = HP3_df['HitposINmmX']
y3 = HP3_df['HitposINmmY']
x4 = HP4_df['HitposINmmX']
y4 = HP4_df['HitposINmmY']

#2D_scatter 생성
ax_scatter.scatter(x1, y1, color = 'black', alpha = 0.3, s = 3, edgecolors = 'none', label = 'sensor1')
ax_scatter.scatter(x2, y2, color = 'green', alpha = 0.3, s = 3, edgecolors = 'none', label = 'sensor2')
ax_scatter.scatter(x3, y3, color = 'red', alpha = 0.3, s = 3, edgecolors = 'none', label = 'sensor3')
ax_scatter.scatter(x4, y4, color = 'blue', alpha = 0.3, s = 3, edgecolors = 'none', label = 'sensor4')

#x_histogram 생성
ax_hist_x.hist(x4, bins = 300, color='blue', alpha=0.3, label = 'sensor4')
ax_hist_x.hist(x3, bins = 300, color='red', alpha=0.3, label = 'sensor3')
ax_hist_x.hist(x2, bins = 300, color='green', alpha=0.3, label = 'sensor2')
ax_hist_x.hist(x1, bins = 300, color='black', alpha=0.3, label = 'sensor1')

#x_cumulative_histogram 생성
ax_Cumhist_x = ax_hist_x.twinx()
ax_Cumhist_x.hist(x1, bins = 100, cumulative = True, histtype = 'step', density = True, color = 'r', align = 'mid')
ax_Cumhist_x.tick_params('y', colors = 'r')

#y_histogram 생성
ax_hist_y.hist(y4, bins = 300, orientation = 'horizontal', color='blue', alpha=0.3, label = 'sensor4')
ax_hist_y.hist(y3, bins = 300, orientation = 'horizontal', color='red', alpha=0.3, label = 'sensor3')
ax_hist_y.hist(y2, bins = 300, orientation = 'horizontal', color='green', alpha=0.3, label = 'sensor2')
ax_hist_y.hist(y1, bins = 300, orientation = 'horizontal', color='black', alpha=0.3, label = 'sensor1')

#y_cumulative_histogram 생성
ax_Cumhist_y = ax_hist_y.twiny()
ax_Cumhist_y.hist(y1, bins = 100, cumulative = True, histtype = 'step', density = True, color = 'r', align = 'mid', orientation = 'horizontal')
ax_Cumhist_y.tick_params('x', colors='r')

#x,y축 scatter 이름 지정
ax_scatter.set_xlabel(xlabel = "X position [mm]", fontsize = 20)
ax_scatter.set_ylabel(ylabel = "Y position [mm]", fontsize = 20)
#x_histogram 이름 지정
ax_hist_x.set_ylabel(ylabel = "Events", fontsize = 20)
ax_Cumhist_x.set_ylabel('cumulative', color = 'r', loc = 'center', fontsize = 20)
#y_histogram 이름 지정
ax_hist_y.set_xlabel(xlabel = "Events", fontsize = 20)
ax_Cumhist_y.set_xlabel('cumulative', color = 'r', loc = 'center', fontsize = 20)

#histogram log scale
ax_hist_x.set_yscale('log')
ax_hist_y.set_xscale('log')

#범례
ax_hist_x.legend(loc = 'center left', fontsize = 17, bbox_to_anchor=(1.07, 0.6))

#이미지를 경로에 저장. 확장자는 .pdf, .png, .jpeg등 가능
#plt.savefig('/Users/ljw/lab/sitrineo/MC_root/figure/Sr90_0.2T_100K_Non_Collimator_2-Dhitposition.png')

plt.show()
```

```python
#3D_scatter
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection = '3d')

#x, y, z축 변수지정
x1 = HP1_df['HitposINmmX']
y1 = HP1_df['HitposINmmY']
z1 = HP1_df['HitposINmmZ']
x2 = HP2_df['HitposINmmX']
y2 = HP2_df['HitposINmmY']
z2 = HP2_df['HitposINmmZ']
x3 = HP3_df['HitposINmmX']
y3 = HP3_df['HitposINmmY']
z3 = HP3_df['HitposINmmZ']
x4 = HP4_df['HitposINmmX']
y4 = HP4_df['HitposINmmY']
z4 = HP4_df['HitposINmmZ']

#3D_scatter 생성
ax.scatter(z4, x4, y4, color = 'blue', alpha = 0.3, s = 3, edgecolors = 'none')
ax.scatter(z3, x3, y3, color = 'red', alpha = 0.3, s = 3, edgecolors = 'none')
ax.scatter(z2, x2, y2, color = 'green', alpha = 0.3, s = 3, edgecolors = 'none')
ax.scatter(z1, x1, y1, color = 'black', alpha = 0.3, s = 3, edgecolors = 'none')

#2D_projection 생성
#sensor4
ax.plot(z4, x4, c = 'blue', alpha=0.5, zdir = 'z', zs = -21, label = 'sensor4') #x, z값을 z축으로 사영
ax.plot(z4, y4, c = 'blue', alpha=0.5, zdir = 'y', zs = 11)                     #y, z값을 y축으로 사영
#sensor3
ax.plot(z3, x3, c = 'red', alpha=0.5, zdir = 'z', zs = -21, label = 'sensor3') 
ax.plot(z3, y3, c = 'red', alpha=0.5, zdir = 'y', zs = 11)
#sensor2
ax.plot(z2, x2, c = 'green', alpha=0.5, zdir = 'z', zs = -21, label = 'sensor2')
ax.plot(z2, y2, c = 'green', alpha=0.5, zdir = 'y', zs = 11)
#sensor1
ax.plot(z1, x1, c = 'black', alpha=0.5, zdir = 'z', zs = -21, label = 'sensor1')
ax.plot(z1, y1, c = 'black', alpha=0.5, zdir = 'y', zs = 11)

#x,y축 이름 지정
ax.set_xlabel("Z position [mm]", fontsize = 20)
ax.set_ylabel("X position [mm]", fontsize = 20)
ax.set_zlabel("Y position [mm]", fontsize = 20)

#x-axis range
ax.set_ylim(-11.0,11.0)
#y-axis range
ax.set_zlim(-21.0,11.0)
#z-axis range
ax.set_xlim(-21.0,25.0)

#범례
plt.legend(loc='upper right', bbox_to_anchor=(1.1, 1.05))

#이미지를 경로에 저장. 확장자는 .pdf, .png, .jpeg등 가능
#plt.savefig('/Users/ljw/lab/sitrineo/MC_root/figure/Sr90_0.2T_100K_Non_Collimator_3-Dhitposition.png')

plt.show()
```
